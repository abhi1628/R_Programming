## **Question:** Performing Linear Regression Analysis in R
You have been given a dataset containing information about various car models, including their weight and fuel efficiency (measured in miles per gallon, MPG). You are required to perform a linear regression analysis to understand the relationship between the weight of the car (wt) and its fuel efficiency (mpg). Follow the steps below to complete this task in R.

1. Load and Explore the Data:
- Load the mtcars dataset from the datasets package.
- Display the first few rows and summary statistics of the dataset.

2. Visualize the Data:
- Create a scatter plot of mpg (miles per gallon) against wt (weight of the car) to visualize the relationship.

3. Fit a Linear Regression Model:
- Fit a linear regression model with mpg as the response variable and wt as the predictor.
- Display a summary of the linear model, which includes coefficients, p-values, R-squared, and residual standard error.

4. Add the Regression Line to the Plot:
- Add the fitted regression line to the scatter plot created in step 2.

5. Model Interpretation:
- Extract and print the intercept and slope of the fitted regression line.
- Predict the MPG for a car that weighs 3.5 (in 1000 lbs) using the fitted model and print the predicted value.

6. Model Diagnostics:
- Plot diagnostic plots to check the assumptions of the linear regression model.
- Extract and print the residual standard error and R-squared value of the model.

Write the R script to perform the above tasks.

```r
# Load necessary library
# The `datasets` package is usually preloaded, but we include it here for completeness.
library(datasets)

# Load the mtcars dataset
data(mtcars)

# Briefly explore the dataset
# View the first few rows of the dataset
head(mtcars)

# Summary statistics of the dataset
summary(mtcars)

# Plot the data to visualize relationships
# Here we plot 'mpg' (miles per gallon) against 'wt' (weight of the car)
plot(mtcars$wt, mtcars$mpg, main="Scatter plot of MPG vs Weight",
     xlab="Car Weight (1000 lbs)", ylab="Miles Per Gallon (MPG)",
     pch=19, col="blue")

# Fit a linear model
# We are predicting 'mpg' using 'wt'
linear_model <- lm(mpg ~ wt, data=mtcars)

# Summary of the linear model
summary(linear_model)

# Adding the regression line to the plot
abline(linear_model, col="red", lwd=2)

# Extracting model coefficients
coefficients <- coef(linear_model)
intercept <- coefficients[1]
slope <- coefficients[2]

# Print the coefficients
cat("Intercept:", intercept, "\n")
cat("Slope:", slope, "\n")

# Predict MPG for a new weight value, e.g., 3.5
new_weight <- 3.5
predicted_mpg <- predict(linear_model, newdata = data.frame(wt = new_weight))

# Print the predicted MPG
cat("Predicted MPG for weight 3.5:", predicted_mpg, "\n")

# Diagnostics: plot residuals
par(mfrow=c(2,2))
plot(linear_model)

# Reset plotting layout
par(mfrow=c(1,1))

# Residual standard error and R-squared value
rse <- summary(linear_model)$sigma
r_squared <- summary(linear_model)$r.squared

# Print the Residual Standard Error and R-squared value
cat("Residual Standard Error:", rse, "\n")
cat("R-squared:", r_squared, "\n")
```
